```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
breast_cancer_data=sklearn.datasets.load_breast_cancer();
b_df=pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
b_df.head()
```

```python
b_df['label'] = breast_cancer_data.target
b_df.tail()
```

```python
b_df.describe()
```

```python
b_df['label'].value_counts()
b_df.groupby('label').mean()
```

Separating the features and the label

```python
X=b_df.drop(columns='label',axis=1)
Y=b_df['label']
```

Splitting the data into training and testing

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
print(X_train)
```

```python
scaler=StandardScaler()
X_train_standard=scaler.fit_transform(X_train)
X_test_standard=scaler.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(30)
from tensorflow import keras
```

```python
model=keras.Sequential([
                          keras.layers.Input(shape=(30,)),
                          keras.layers.Flatten(),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
                      ])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Training the Neural Network

```python
history=model.fit(X_train_standard,Y_train,validation_split=0.1,epochs=15)
```

```python
loss, accuracy = model.evaluate(X_test_standard, Y_test)
print(accuracy)
```

```python
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

plt.plot(train_accuracy, label='Train Accuracy')
plt.plot(val_accuracy, label='Validation Accuracy')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.show()
```

```python
loss=history.history["loss"]
val_loss=history.history["val_loss"]
plt.plot(loss,label="Loss")
plt.plot(val_loss,label="Val_loss")
plt.title("Loss Compare")
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(X_test_standard)
Y_pred_labels = [np.argmax(i) for i in y_pred]
print(Y_pred_labels)
```
